```python
def factorial(n):
    """Calculate the factorial of n."""
    if n == 0:
        return 1
    else:
        result = 1
        for i in range(1, n + 1):
            result *= i
        return result

def radians(degrees):
    """Convert degrees to radians."""
    return degrees * (3.141592653589793 / 180.0)

def sine_taylor_series(x, terms=10):
    """Calculate sine using the Taylor series expansion."""
    sine = 0
    for n in range(terms):
        sign = (-1) ** n
        sine += sign * (x ** (2 * n + 1)) / factorial(2 * n + 1)
        print("n =", n)
        print ("sign =", sign )
        print( "sine ==", sine)
        print( "factorial(2 * n + 1) ==", factorial(2 * n + 1))

    return sine

def cosine_taylor_series(x, terms=10):
    """Calculate cosine using the Taylor series expansion."""
    cosine = 0
    for n in range(terms):
        sign = (-1) ** n
        cosine += sign * (x ** (2 * n)) / factorial(2 * n)
    return cosine

# Input: angle in degrees
angle_degrees = float(input("Enter the angle in degrees: "))

# Convert angle to radians
angle_radians = radians(angle_degrees)

# Calculate sine and cosine using Taylor series
sine = sine_taylor_series(angle_radians)
cosine = cosine_taylor_series(angle_radians)

# Output the results
print(f"Sine of {angle_degrees} degrees is: {sine}")
print(f"Cosine of {angle_degrees} degrees is: {cosine}")
```

```python
%matplotlib widget
import matplotlib as mpl
import matplotlib.pyplot as plt
import numpy as np

D= 0.1 ##mm
print("Diameter (m)=", D)
A= np.pi* D**2 /4
print("Area =" , A)
flow= 32/3600 # m^3/sec ~60L/h
print("FLOW rate (m^3/s)=" , flow)

# air velocity
V= flow/A
print( "V speed air flow (m/s)= " , V)

############################################
rho_air= 1.225 #kg/m3
mu_dynamic = 1.81e-5 #kg/ms
Re_criti = 2300

d_critic = Re_criti*mu_dynamic / ( rho_air * V  )
print("Reynolds critical length (m)=", d_critic)



################################
################################
# for x< d_critic
X_array_s = np.arange(0,d_critic,0.0001)
# print("X_array_s =",X_array_s)

v_x_s = flow/(A*X_array_s) # SI
v_x_s = np.tile(flow/A , len(X_array_s) )
# v_x_s = (flow/A)*len(X_array_s)
print("air velocity v_x_s (m/s)= ",v_x_s)
f_v_x_s = v_x_s*A # SI
# print("f_v_x_s = ",f_v_x_s)


################################
###############################
#  for x> d_critic
X_array = np.arange(d_critic,1,0.0001)
# Calculate The velosity at the critical disence
v_re_critical = Re_criti * mu_dynamic / (d_critic * rho_air)

print("v_re_critical = ", v_re_critical)

# air velocity
v_x_l = v_re_critical*(d_critic/X_array)**1.5
print(v_x_l)

# flow rate m^3/s
f_v_x_l = v_x_l *A
print("air flow through the same area (m^3/s)", f_v_x_l)
print("air flow through the same area (L/min)", f_v_x_l*1e3*60)


a=X_array_s
b=v_x_s
c=X_array
d=v_x_l
# Create a figure and a set of subplots
fig, ax = plt.subplots(figsize=(8, 6))

# Plot the data
ax.plot(a, b, marker='o', linestyle='-', color='g')
ax.plot(c, d, marker='o', linestyle='-', color='r')

# Adding titles and labels
ax.set_title('Air velocity')
ax.set_xlabel('Distance (m)')
ax.set_ylabel('air velocity m/s')

# Add a grid
ax.grid(True)

# Show the plot
plt.show()
```
